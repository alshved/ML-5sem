Сделайте кластеризацию данных методом `KMeans` (собственная имплементация и готовая реализация), обучив модель на данных со сниженной за счет PCA размерностью. В данном случае мы подскажем, что нужно искать именно 6 кластеров, но в общем случае мы не будем знать, сколько кластеров надо искать.

Параметры:

- **n_clusters** = n_classes (число уникальных меток целевого класса)
- **n_init** = 100
- **random_state** = RANDOM_STATE (для воспроизводимости результата)

Остальные параметры со значениями по умолчанию.

````python
class MyKMeans:
    def __init__(self, n_clusters=6, n_init=10, max_iter=300, random_state=None):
        self.n_clusters = n_clusters
        self.n_init = n_init
        self.max_iter = max_iter
        self.random_state = random_state
        self.cluster_centers_ = None
        self.labels_ = None
        self.inertia_ = None

    def fit(self, X):
        np.random.seed(self.random_state)
        best_inertia = np.inf
        
        for _ in range(self.n_init):
            indices = np.random.permutation(X.shape[0])[:self.n_clusters]
            centroids = X[indices]
            
            for _ in range(self.max_iter):
                distances = np.linalg.norm(X[:, np.newaxis] - centroids, axis=2)
                labels = np.argmin(distances, axis=1)
                
                new_centroids = np.array([X[labels == k].mean(axis=0) 
                                          if np.sum(labels == k) > 0 else centroids[k] 
                                          for k in range(self.n_clusters)])
                
                if np.allclose(centroids, new_centroids):
                    break
                    
                centroids = new_centroids
            
            current_inertia = 0
            for k in range(self.n_clusters):
                cluster_points = X[labels == k]
                if len(cluster_points) > 0:
                    current_inertia += np.sum((cluster_points - centroids[k])**2)
            
            if current_inertia < best_inertia:
                best_inertia = current_inertia
                self.cluster_centers_ = centroids
                self.labels_ = labels
                self.inertia_ = best_inertia
                
        return self
````

````python
sklearn_kmeans = KMeans(n_clusters=6, n_init=100, random_state=RANDOM_STATE)
sklearn_kmeans.fit(X_pca)

my_kmeans = MyKMeans(n_clusters=6, n_init=100, random_state=RANDOM_STATE)
my_kmeans.fit(X_pca)

print(f"Sklearn Inertia: {sklearn_kmeans.inertia_:.2f}")
print(f"MyKMeans Inertia: {my_kmeans.inertia_:.2f}")
````
